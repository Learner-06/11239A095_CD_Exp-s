```python
def is_keyword(word, keywords):
 return word.upper() in keywords
keywords1 = {"ORDER", "CANCEL", "TRACK"}
print(is_keyword("TRACK", keywords1)) # True
keywords2 = {"ISSUE", "RETURN", "SEARCH"}
print(is_keyword("RETURN", keywords2)) # True
```
