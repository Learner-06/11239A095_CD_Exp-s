```python
import ply.lex as lex
import sys

# Workaround for ply in interactive environments: ensure __main__ module has a __file__ attribute
if not hasattr(sys.modules['__main__'], '__file__'):
    sys.modules['__main__'].__file__ = 'lexer_rules.py'

reserved = {'ADD': 'ADD'}
tokens = ['ID', 'NUMBER'] + list(reserved.values())

def t_ID(t):
    r'[a-zA-Z_][a-zA-Z0-9_]*'
    t.type = reserved.get(t.value, 'ID')
    return t

def t_NUMBER(t):
    r'\d+'
    t.value = int(t.value)
    return t

t_ignore = ' \t'

def t_error(t):
    print("Illegal character:", t.value[0])
    t.lexer.skip(1)

# Fix: Pass the __main__ module directly to ply.lex()
# This uses the __file__ attribute we set on sys.modules['__main__']
lexer = lex.lex(module=sys.modules['__main__'])

lexer.input("ADD item1 2")

for tok in lexer:
    print(tok)
```
